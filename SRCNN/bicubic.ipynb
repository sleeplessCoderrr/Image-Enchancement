```python
from pathlib import Path
from PIL import Image
```

```python
def bicubic_resize(input_path: str, output_path: str, scale: float = 2.0):
    src = Path(input_path)
    dst = Path(output_path)
    dst.parent.mkdir(parents=True, exist_ok=True)

    with Image.open(src) as img:
        new_w = int(img.width * scale)
        new_h = int(img.height * scale)
        resized = img.resize((new_w, new_h), resample=Image.BICUBIC)
        resized.save(dst)
    print(f"Saved bicubic image to {dst}")

if __name__ == "__main__":
    bicubic_resize("dataset/DIV2K_train_LR_bicubic/X4/0002x4.png", "output/bicubic_output2.jpg", scale=2.0)
```
